```
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

data = pd.DataFrame()
apps = ["chrome", "youtube", "snapchat", "gmail", "messenger"]
usage = [25, 18, 16, 10, 8]
data['apps'] = apps
data['usage'] = usage
plt.figure(figsize=(6,6))
plt.pie(usage, labels=data['apps'], shadow=True, autopct='%1.0f%%')
plt.title("phone usage")
plt.show()
```
